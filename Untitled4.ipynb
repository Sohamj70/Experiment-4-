# **Addition and removal of elements in a set**


```python
thisset = {"apple", "banana", "cherry"}

thisset.add("orange")
thisset.add(2)
thisset.remove("banana")

print(thisset)

#we cannot replace a item in a set but we can add and remove a element
```

# ***Set Operations***


```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Union : ",A|B)
print("intersection : " ,A & B)
print("Difference1 : ",A - B)
print("Difference2 : ",A ^ B)
print("Symmetric Difference : ",A ^ B)
```

# ***FrozenSet***
